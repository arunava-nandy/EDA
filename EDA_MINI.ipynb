```python
#Replacing the symbols
```

```python
df1['Value']=df1['Value'].str.replace('€','')
df1['Value']=df1['Value'].str.replace('M','')
df1['Value']=df1['Value'].str.replace('K','')
df1['Wage']=df1['Wage'].str.replace('€','')
df1['Wage']=df1['Wage'].str.replace('M','')
df1['Wage']=df1['Wage'].str.replace('K','')
df1['Height']=df1['Height'].str.replace("'",'.')
df1['Weight']=df1['Weight'].str.replace("lbs",'')
df1['Release Clause']=df1['Release Clause'].str.replace('€','')
df1['Release Clause']=df1['Release Clause'].str.replace('M','')
df1['Release Clause']=df1['Release Clause'].str.replace('K','')
```

```python
for i in df1['Joined']:
    df1['Joined']=i[-4:]
```

```python
#Converting DataType
```

```python
df1['Value']=pd.to_numeric(df1['Value'],errors='coerce')
df1['Wage']=pd.to_numeric(df1['Wage'],errors='coerce')
df1['Joined']=pd.to_numeric(df1['Joined'],errors='coerce')
df1['Height']=pd.to_numeric(df1['Height'],errors='coerce')
df1['Weight']=pd.to_numeric(df1['Weight'],errors='coerce')
df1['Release Clause']=pd.to_numeric(df1['Release Clause'],errors='coerce')
```

```python
df1['Contract Valid Until'] = pd.to_datetime(df1['Contract Valid Until'])
```

```python
#Check Null Values
```

```python
df1.isna().sum()
```

```python
#Null value imputation
```

```python
df1=df1.fillna(df1.mean())
```
